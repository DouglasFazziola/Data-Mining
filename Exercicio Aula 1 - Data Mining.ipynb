## Preparando Dataframe

## Criando Coluna "Faixa de Renda"

```python
df["Faixa de Renda"] = np.nan
```

```python
df.head()
```

```python
def faixarenda(registro1):
    if registro1['Renda'] < 1000:
        return 'menor que 1000'
    elif registro1['Renda'] >= 1000 and registro1['Renda'] < 1600:
        return '1000|- 1600'
    elif registro1['Renda'] >= 1600 and registro1['Renda'] < 4400:
        return '1600|-4400'
    elif registro1['Renda'] >= 4400 and registro1['Renda'] < 5400:
        return '4400|-5400'
    elif registro1['Renda'] >= 5400:
        return 'Maior que 5400'
    else:
        return 'n/a'
```

```python
df['Faixa de Renda'] = df.apply(faixarenda, axis=1)
```

```python
df.head()
```

```python
set(df['Faixa de Renda'])
```

## Criando Coluna "Classe"

```python
df["Classe"] = np.nan
df.head()
```

```python
def classe(registro2):
    if registro2['Renda'] <= 2200:
        return 'E'
    if registro2['Renda'] > 2200 and registro2['Renda'] <= 4400:
        return 'D'
    if registro2['Renda'] > 4400 and registro2['Renda'] <= 11000:
        return 'C'
    if registro2['Renda'] > 11000 and registro2['Renda'] <= 22000:
        return 'B'
    if registro2['Renda'] > 22000:
        return 'A'
    else:
        return 'n/a'
```

```python
df['Classe'] = df.apply(classe, axis=1)
```

```python
df.head()
```

```python
set(df['Classe'])
```

## Adicionar Coluna "Faixa Etária"

```python
df["Faixa Etária"] = np.nan
df.head()
```

```python
def faixa(registro3):
    if registro3['idade'] >= 16 and registro3['idade'] <= 24:
        return '16 a 24'
    if registro3['idade'] > 24 and registro3['idade'] <= 34:
        return '24 a 34'
    if registro3['idade'] > 34 and registro3['idade'] <= 44:
        return '34 a 44'
    if registro3['idade'] > 44 and registro3['idade'] <= 54:
        return '44 a 54'
    if registro3['idade'] > 54 and registro3['idade'] <= 64:
        return '54 a 64'
    if registro3['idade'] >= 65:
        return '65+'
    else:
        return 'n/a'
```

```python
df['Faixa Etária'] = df.apply(faixa, axis=1)
```

```python
df.head()
```

```python
set(df['Faixa Etária'])
```

```python
df.loc[df['Faixa Etária'] == 'n/a']
```

## Identificando "Sexo"

```python
def sexo(registro4):
    if registro4['Sexo'] == '0':
        return 'M'
    elif registro4['Sexo'] == '1':
        return 'F'
    else:
        return 'D'
```

```python
df['Sexo'] = df.apply(sexo, axis=1)
```
